```python
cur.execute('''
CREATE TABLE alumnos AS
SELECT Nombre, Email, Promoción, Fecha_comienzo, Campus, 
       Proyecto_HLF, Proyecto_EDA, Proyecto_BBDD, Proyecto_ML, Proyecto_Deployment,
       NULL AS Proyecto_WebDev,
       NULL AS Proyecto_FrontEnd,
       NULL AS Proyecto_Backend,
       NULL AS Proyecto_React,
       NULL AS Proyecto_FullSatck
FROM union1

UNION ALL

SELECT Nombre, Email, Promoción, Fecha_comienzo, Campus, 
       NULL AS Proyecto_HLF,
       NULL AS Proyecto_EDA,
       NULL AS Proyecto_BBDD,
       NULL AS Proyecto_ML,
       NULL AS Proyecto_Deployment,
       Proyecto_WebDev,
       Proyecto_FrontEnd,
       Proyecto_Backend,
       Proyecto_React,
       Proyecto_FullSatck
FROM union2;
''')
# Guardar cambios
con.commit()
```

```python
# Con esta función leemos los datos y lo pasamos a un DataFrame de Pandas
def sql_query(query):

    # Ejecuta la query
    cur.execute(query)

    # Almacena los datos de la query 
    ans = cur.fetchall()

    # Obtenemos los nombres de las columnas de la tabla
    names = [description[0] for description in cur.description]

    return pd.DataFrame(ans,columns=names)
```

```python
cur.execute('''
ALTER TABLE alumnos ADD COLUMN alumno_id TEXT;
''')

# Actualizar la columna con valores '001', '002', ...
cur.execute('''
WITH numbered AS (
    SELECT rowid, printf('%03d', rowid) AS id_str
    FROM alumnos
)
UPDATE alumnos
SET alumno_id = (SELECT id_str FROM numbered WHERE numbered.rowid = alumnos.rowid);
''')

con.commit()
```

```python
query = '''
SELECT *
FROM alumnos
;
'''
display(sql_query(query))
```

```python
cur.execute('''
ALTER TABLE claustro ADD COLUMN profesor_id TEXT;
''')

# Actualizar la columna con valores '01', '02', ...
cur.execute('''
WITH numbered1 AS (
    SELECT rowid, printf('%02d', rowid) AS id_str
    FROM claustro
)
UPDATE claustro
SET profesor_id = (SELECT id_str FROM numbered1 WHERE numbered1.rowid = claustro.rowid);
''')

con.commit()
```

```python
query = '''
SELECT *
FROM claustro
;
'''
display(sql_query(query))
```
